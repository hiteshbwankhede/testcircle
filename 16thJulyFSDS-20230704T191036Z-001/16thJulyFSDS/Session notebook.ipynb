# Remove HTML Tags

```python
text = """<div>
<h1>Title</h1>
<p>A long text........ </p>
<a href=""> a link </a>
</div>"""
```

```python
import re
# as per recommendation from @freylis, compile once only
CLEANR = re.compile('<.*?>') 

def cleanhtml(raw_html):
  cleantext = re.sub(CLEANR, '', raw_html)
  return cleantext
```

```python
cleanhtml(text)
```

# Tokenization

```python

mytext = '''In the previous chapter, we saw examples of some common NLP 
applications that we might encounter in everyday life. If we were asked to 
build such an application, think about how we would approach doing so at our 
organization.'''
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
import nltk
nltk.download('punkt')
```

```python
my_sentences = sent_tokenize(mytext)
print(my_sentences)
```

```python
# words tokenization
for sentence in my_sentences:
   print(sentence,"\n")
   print(word_tokenize(sentence),"\n")
```
